```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados
dados = {
    'JAN': [7.50, 2.43, 3.06],
    'FEV': [4.56, 2.51, 3.06],
    'MAR': [2.45, 3.04, 3.06]
}

# Índices e colunas
indices = ['Home', 'Presencial', 'Meta']
colunas = ['JAN', 'FEV', 'MAR']

# Criando o DataFrame
df = pd.DataFrame(dados, index=indices, columns=colunas)

# Plotando o gráfico
plt.figure(figsize=(10, 6))

# Plotando as barras para os dados de Home e Presencial
df.loc[['Home', 'Presencial']].T.plot(kind='bar', ax=plt.gca(), color=['blue', 'green'])
# Plotando a linha para a meta
df.loc['Meta'].plot(ax=plt.gca(), color='red', linestyle='--', linewidth=2)

plt.title('Comparativo de Home, Presencial e Meta')
plt.xlabel('Mês')
plt.ylabel('Porcentagem')
plt.xticks(rotation=0)
plt.legend(['Home', 'Presencial', 'Meta'])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
quartis = ['Q1', 'Q2', 'Q3', 'Q4']
jan = [344, 431, 534, 684]
fev = [383, 506, 640, 880]

# Configuração do gráfico
x = np.arange(len(quartis))
width = 0.35

fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, jan, width, label='Jan')
bars2 = ax.bar(x + width/2, fev, width, label='Fev')

# Adicionando rótulos
ax.set_xlabel('Quartil')
ax.set_ylabel('Evolução')
ax.set_title('Evolução por Quartil e Mês')
ax.set_xticks(x)
ax.set_xticklabels(quartis)
ax.legend()

# Adicionando o valor de cada barra
def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(bars1)
autolabel(bars2)

# Mostrar o gráfico
plt.grid(True)
plt.tight_layout()
plt.show()

```
